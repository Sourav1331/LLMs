**A token embedding is a vector (list of numbers) that represents a token (word or part of a word) in a way that captures its meaning.**

**Since computers can’t understand words directly, we convert each token into numbers → that numeric form is the embedding.**

**Example:**
**Sentence:   "I love apples"**

**Tokenizer → tokens:**
**["I", "love", "apple", "s"] (sometimes words are split)**

**Each token becomes a vector:**

*"I"     → [0.12, -0.55, 1.03, ..., 0.07]*

*"love"  → [-0.9,  0.34, 0.88, ..., 1.21]*

*"apple" → [0.44,  1.02, -0.33, ..., 0.15]*

*"s"     → [ ... ]*
